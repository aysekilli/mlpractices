```python
from sklearn.linear_model import LogisticRegression

X_train = [[0, 0], [1, 1]]                        #eğitim verisi, iki özellikten oluşan iki örnek
y_train = [0, 1]                                  #hedef değişken, iki örneğin sırasıyla 0 ve 1 sınıflarına ait olduğunu gösteriyor
X_test = [[2, 2], [3, 3], [4, 5], [5, 6],[-1, 0]] #test verisi


model = LogisticRegression()            #Logistic Regression modelini oluşturduk
model.fit(X_train, y_train)             #modeli x_train ve y_train verileriyle eğittik
predictions = model.predict(X_test)     #modeli kullanarak x_test verisi üzerinde tahminler yaptık ve sonuçları predictions değişkenine atadık
print(predictions)
```
