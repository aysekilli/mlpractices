```python
#ona en yakın k komşusuna bakarak çoğunluğu tahmin ediyo

from sklearn.neighbors import KNeighborsClassifier #KNN algoritmasını kullanmak için gerekli kütüphane
from sklearn.datasets import load_iris #iris veri setini yüklemek için gerekli kütüphane
import numpy as np #numpy kütüphanesini kullanmak için


df = load_iris()     #veri setini yükledik

X = df.data         #veri setindeki featureları X değişkenine atadık
y = df.target       #veri setindeki hedef değişkeni y değişkenine atadık

print(X.shape)      #X değişkeninin boyutunu yazdırdık
print(y.shape)      #y değişkeninin boyutunu

knn = KNeighborsClassifier(n_neighbors=3) #KNN algoritmasını oluşturduk ve k değerini 3 olarak belirledik
knn.fit(X, y)       #mesela niye büyük X ? = çünkü sklearn'de genellikle büyük X featureları temsil eder, küçük y ise hedef değişkeni temsil eder. Bu sadece bir konvansiyon, ama genellikle bu şekilde kullanılırmış.
                    #nedir bu fit? = train.

X_new = np.array([[5.0, 3.0, 1.0, 0.2]])    # yeni bir veri noktası oluşturduk
print(X_new.shape)  # veri noktasının boyutunu yazdırdık

prediction = knn.predict(X_new)  # yeni veri noktasını kullanarak tahmin
print(prediction)

print(df.target_names[prediction]) #tahmin edilen sınıfın adını yazdırdık
```
